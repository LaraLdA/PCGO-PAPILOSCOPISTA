```python
def show_info(data):
    print('DATASET SHAPE: ', data.shape, '(linhas e colunas) \n')
    print('-'*50)
    print('FEATURE DATA TYPES:')
    print('NUMBER OF UNIQUE VALUES PER FEATURE:', '\n')
    print(data.nunique())
    print(data.info())
```

```python
papiloscopista.drop(columns=['1.60','1.20','2.00','1.20.1','1.20.2','3.20','1.60.1','1.20.3','1.60.2','0.40','1.20.4','0.00','0.80'], inplace=True)
```

```python
# acrescentando ao final do dataframe o candidato que ficou como cabeçalho da tabela
papiloscopista.loc[len(papiloscopista)]=['Abner Lopes Dantas','3280013924','17.20'] 
```

```python
show_info(papiloscopista)
```

```python
#renomeando colunas
papiloscopista.columns=['NOME','INSCRIÇÃO','NOTA OBJETIVA']
```

```python
papiloscopista.head()
```

```python
#str para float
papiloscopista['NOTA OBJETIVA'] = pd.to_numeric(papiloscopista['NOTA OBJETIVA'])
papiloscopista.sort_values(by='NOTA OBJETIVA',ascending= False)
```

```python
#organizando por nota e alfabeto
papi_organizado = papiloscopista.sort_values(by=['NOTA OBJETIVA','NOME'],ascending= [False,True]).reset_index(drop=True)
```

```python
papi_organizado.describe()
```

```python
#Pegando apenas as notas iguais ou maiores a 22
Notas_apartir_22 = papi_organizado.loc[(papi_organizado['NOTA OBJETIVA'] >= 22.0)]
```

```python
Notas_apartir_22
```

```python
import matplotlib
```

```python
#Colocando cores diferentes para cada valor das notas
colors = dict(zip(Notas_apartir_22['NOTA OBJETIVA'].unique(),
                  (f'background-color: {c}' for c in matplotlib.colors.cnames.values())))

firts500=Notas_apartir_22.style.applymap(colors.get, subset=['NOTA OBJETIVA'])
```

```python
primeiros500 = firts500.set_table_attributes('border=1 cellspacing=0 text-align:center').set_properties(**{'width': '200px', 'color': 'red','border-color': 'grey','text-align':'center',
                                                                                                           'font-weight':'bold'},subset=['NOTA OBJETIVA']).set_precision(2).set_caption("PCGO- PAPILOSCOPISTA (2023)| 500 PRIMEIROS [BASEADO NO RESULTADO DA PROVA OBJETIVA PRELIMINAR]")

testando = primeiros500.render()
with open("primeiros500.html","w") as fp:
    fp.write(testando)
```

```python
from IPython.display import IFrame
IFrame("Primeiros500.html", width=600, height=300)
```

```python
todos = papi_organizado.style.set_table_attributes('border=1 cellspacing=0 text-align:center').set_properties(**{'width': '200px', 'color': 'black','border-color': 'grey','text-align':'center','font-weight':'bold',
                                        'background-color': '#adff92'},subset=['NOTA OBJETIVA']).set_precision(2).set_caption("PCGO- PAPILOSCOPISTA (2023)| CLASSIFICACAO GERAL (BASEADO NO RESULTADO DA PROVA OBJETIVA PRELIMINAR)")

testando = todos.render()
with open("todos.html","w") as fp:
    fp.write(testando)
```
